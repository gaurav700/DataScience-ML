# Working with missing data

```python
data.isnull().sum()
```

```python
data['Gender'] = data['Gender'].replace(np.nan, 'M')
```

```python
data.isnull().sum()
```

```python
data['Prefix'] = data['Prefix'].bfill() # Backward fill means it fills the nan value with value below that and in forward fill its vice versa.
```

```python
data.isnull().sum()
```

# Column transformations

```python
data.head()
```

```python
# where children are 2 we give them 10000 bonus and if more than that we give 0 and if less than 2 we give them 20000 bonus

data.loc[(data['TotalChildren'] == 2), "Bonus"] = "$10000"
data.loc[(data['TotalChildren'] < 2), "Bonus"] = "$20000"
data.loc[(data['TotalChildren'] > 2), "Bonus"] = "$0"
```

```python
data
```

```python
data['FirstName'] = data['FirstName'].str.capitalize() + " " + data['LastName'].str.capitalize()
```

```python
data = data.drop(columns=['LastName'])
```

```python
data.rename(columns={'FirstName': 'FullName'}, inplace=True)
```

```python
data['FullName'] = data['Prefix'] + " " + data['FullName']
data = data.drop(columns=['Prefix'])
```

```python
data.loc[data['MaritalStatus'] == 'M', 'MaritalStatus'] = 'Married'
data.loc[data['MaritalStatus'] == 'S', 'MaritalStatus'] = 'Single'
```

```python
data
```

# Map using

```python
months = {"Months": ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']}
df = pd.DataFrame(months)
```

```python
def extract(value):
  if(value =='September'):
    return value[:4].upper()
  return value[:3].upper()

df['Abbreviations'] = df['Months'].map(extract)
```

```python
df
```
